```python
from surprise import SVD
from surprise import Dataset
from surprise.model_selection import GridSearchCV
```

Завантаження датасету movielens

```python
data = Dataset.load_builtin('ml-100k')
```

Модель SVD

```python
svd = SVD()

param_grid = {'n_epochs': [5, 10], 'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}
grid_search = GridSearchCV(SVD, param_grid, measures=['RMSE'], cv=3)
grid_search.fit(data)

print('Найкращі параметри:', grid_search.best_params)
print('Найкраща оцінка RMSE:', grid_search.best_score['rmse'])
```

Зразок використання оптимальних параметрів

```python
best_params = grid_search.best_params['rmse']
svd = SVD(n_epochs=best_params['n_epochs'], lr_all=best_params['lr_all'], reg_all=best_params['reg_all'])
```

Оцінка моделі за допомогою крос-валідації

```python
from surprise.model_selection import cross_validate

cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

SVD++

```python
from surprise import SVDpp

param_grid_svdpp = {'n_epochs': [5, 10], 'lr_all': [0.002, 0.005],
                    'reg_all': [0.4, 0.6], 'n_factors': [20, 50]}

grid_search_svdpp = GridSearchCV(SVDpp, param_grid_svdpp, measures=['RMSE'], cv=3)
grid_search_svdpp.fit(data)

print('Найкращі параметри для SVD++:', grid_search_svdpp.best_params)
print('Найкраща оцінка RMSE для SVD++:', grid_search_svdpp.best_score['rmse'])
```

NMF

```python
from surprise import NMF

param_grid_nmf = {'n_epochs': [10, 20, 30], 'lr_bu': [0.002, 0.005],
                  'reg_bu': [0.4, 0.6], 'reg_bi': [0.4, 0.6]}

grid_search_nmf = GridSearchCV(NMF, param_grid_nmf, measures=['RMSE'], cv=3)
grid_search_nmf.fit(data)

print('Найкращі параметри для NMF:', grid_search_nmf.best_params)
print('Найкраща оцінка RMSE для NMF:', grid_search_nmf.best_score['rmse'])
```

Результати показують, що алгоритм SVD має менше значення RMSE порівняно з NMF та SVD++. Таким чином, на основі цих даних можна припустити, що SVD є більш оптимальним алгоритмом для даного датасету.
